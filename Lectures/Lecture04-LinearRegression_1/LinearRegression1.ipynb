```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn
seaborn.set_style('darkgrid')
```

```python
example_data = pd.read_csv('example1.csv')
```

```python
D = example_data[['x','y']].to_numpy()
D.shape
```

```python
mhat, bhat = np.polyfit(D[:,0], D[:,1], 1)
```

```python
plt.scatter(D[:,0], D[:,1], edgecolors='xkcd:midnight blue')
plt.plot([0, 2], [mhat*0+bhat, mhat*2+bhat], c='crimson')
plt.show()
```

```python
mse = np.sum((D[:,1] - mhat*D[:,0] - bhat)**2)/len(D[:,0])
```
